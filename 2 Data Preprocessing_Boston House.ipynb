# Splitting data into train and test Datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bostonhouse.data, bostonhouse.target, test_size=0.20, random_state=0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Feature Scaling

```python
#Standardizing Data
from sklearn.preprocessing import StandardScaler 
sc_x=StandardScaler()
x_train_ss=sc_x.fit_transform(X_train)
x_test_ss=sc_x.transform(X_test)

pdstdscaled = pd.DataFrame(x_train_ss)
print(pdstdscaled.shape)
print(pdstdscaled.head(10))
pdstdscaled.describe()
```

```python
pdstdscaled.hist()
plt.show()
```

```python
#Normalizing the data
from sklearn.preprocessing import Normalizer
scaler=Normalizer()
x_train_normalizer=scaler.fit_transform(X_train)
x_test_normalizer=scaler.transform(X_test)

pdnormalscaled = pd.DataFrame(x_train_normalizer)
print(pdnormalscaled.shape)
print(pdnormalscaled.head(10))
pdnormalscaled.describe()
```

```python
#Rescaling Data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-1,1))
x_train_mm=scaler.fit_transform(X_train)
x_test_mm=scaler.transform(X_test)

pdscaled = pd.DataFrame(x_train_mm)
print(pdscaled.shape)
print(pdscaled.head(10))
#pdscaled.describe()
```

```python
#Binarizing Data
from sklearn.preprocessing import Binarizer
binarizer=Binarizer()
x_train_binarizer=binarizer.fit_transform(X_train)
x_test_binarizer=binarizer.transform(X_test)

pdbinarized = pd.DataFrame(x_train_binarizer)
print(pdbinarized.shape)
print(pdbinarized.tail(10))
```
